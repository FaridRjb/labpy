```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
```

```python
class Projectile:
    
    def __init__(self, x0, y0, vx0, vy0):
        self.x = x0
        self.y = y0
        self.vx = vx0
        self.vy = vy0
    
    def update(self, new_x, new_y, new_vx, new_vy):
        self.x = new_x
        self.y = new_y
        self.vx = new_vx
        self.vy = new_vy
        
    def __str__(self):
        return 'x=' + str(self.x) + ', y=' + str(self.y) + ', vx=' + str(self.vx) + ', vy=' + str(self.vy)
```

```python
distri = True
```

```python
if distri:
    ps = [Projectile(-0.1, rnd.uniform(0, 0.05), 10, 0) for _ in range(5000)]
else:
    ps = [Projectile(-0.1, i, 10, 0) for i in np.linspace(0, 0.025, 11)]
    
    fig, axis = plt.subplots(dpi=150)
    axis.scatter(0, 0, c='blue', s=20)
    
    axis.set_xlim(-0.1, 0.1)
    axis.set_ylim(-0.025, 0.1)
    axis.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
    axis.set_aspect('equal', adjustable='box')
    axis.grid(axis='y')
    
    colors = ['gray', 'rosybrown', 'brown', 'red', 'orange', 'gold',
              'yellowgreen', 'lime', 'aqua', 'skyblue', 'blueviolet',
              'plum', 'hotpink']

K = 1

max_step = 200
dt = 1e-4

for i, p in enumerate(ps):
    for _ in range(0, max_step):
        r2 = (p.x - 0)**2 + (p.y - 0)**2
        a = -K / r2
        alpha = np.arctan((0 - p.y) / (0 - p.x))
        ax = a * np.cos(alpha)
        ay = a * np.sin(alpha)
        new_x = p.x + p.vx * dt + 0.5 * ax * dt**2
        new_y = p.y + p.vy * dt + 0.5 * ay * dt**2
        new_vx = p.vx + ax * dt
        new_vy = p.vy + ay * dt
        
        if not distri: axis.scatter(new_x, new_y, c=colors[i], s=1)
        
        p.update(new_x, new_y, new_vx, new_vy)

if not distri:
    fig.savefig('RS.png')
```

```python
def phi(projectile):
    theta = np.arctan(projectile.vy / projectile.vx) * 180 / np.pi
    if theta <= 0: theta += 180
    return theta

data = list(map(phi, ps))

fig, axis = plt.subplots(dpi=150)
axis.hist(data, color='lightgreen', ec='black', bins=180)
axis.set_xlim(0, 180)
axis.set_xlabel('Deflection angle (Degree)')
axis.set_ylabel('Population')
axis.tick_params(left=False, labelleft=False)
fig.savefig('RS.png')
```
